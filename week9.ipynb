```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

# First let's try the most basic way to do it: store the results into a 2D array

```python
# First estimate what range of nx ,ny, nz should we use?
Emax = 60
nmax = int(np.floor(np.sqrt(Emax-1-1)))
results = []
```

```python
nmax
```

```python
for nx in range(1,nmax+1):
    for ny in range(1,nmax+1):
        for nz in range(1,nmax+1):
            E = nx*nx + ny*ny + nz*nz
            if E <= Emax:
                results.append([nx,ny,nz,E])
            
```

```python
results
```

```python
results=np.array(results)
```

```python
results
```

```python
sorted_results = results[np.argsort(results[:,3])]
```

```python
sorted_results
```

# Now let's try something fancy: pandas dadtaframe

```python
import pandas as pd
```

```python
df = pd.DataFrame(data=results,columns=['nx','ny','nz','E'])
```
