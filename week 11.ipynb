```python
import matplotlib.pyplot as plt
from matplotlib import cm, colors
import numpy as np
from scipy import special
```

```python
npoints=200
xlower = -2
xupper = 10
ylower = -5
yupper = 10

# Define the figure size
fig = plt.figure(figsize=(6,4))

# Give the range of x values to plot 
x = np.linspace(xlower,xupper,npoints)

# Add a title
fig.suptitle('Laguerre polynomials')

# Show the grid lines
plt.grid(b=True)

# Define the colors and line styles to draw
colors = ['c','b','r','g']
styles = ['-','--','-.',':']

# Loop over different polynomials
for n in range(0,4):
    for alpha in range(0,n+1):
        L=special.genlaguerre(n,alpha)
        label=r'$L_'+ str(n)+r'^{('+ str(alpha) + r')}$'
        print(label)
        plt.plot(x,L(x),colors[n]+styles[alpha], label=label)

# Set the range of x
plt.xlim(xlower,xupper)
plt.ylim(ylower,yupper)

# Add a Legend
plt.legend(ncol=3, fontsize=8)

# Show the plot
```

```python
# Now plot the Radial functions
import math
```

```python
npoints=200
xlower = 0
xupper = 15
ylower = -0.1
yupper = 2

# Define the figure size
fig, axs = plt.subplots(1,3,sharex=True,sharey=True,figsize=(12,6))


# Give the range of x values to plot 
x = np.linspace(xlower,xupper,npoints)

# Add a title
fig.suptitle('Radial functions of hydrogen atom')


# Define the colors and line styles to draw
colors = ['c','b','r','g']
styles = ['-','--','-.',':']

# Loop over different polynomials
for n in range(1,4):
    for l in range(0,n):
        coeff= math.sqrt(math.factorial(n-l-1)/(2*n*math.factorial(n+l)))*(2/float(n))**(l+1.5)
        L=special.genlaguerre(n-l-1,2*l+1)
        R = coeff*(x**l)*np.exp(-x/n)*L(2*x/n)
        label=r'$R_{'+ str(n)+r','+ str(l) + r'}$'
        print(label)
        axs[n-1].plot(x,R,colors[n]+styles[l], label=label)
    # Set the upper and lower bounds for diplaying the figur
    print(yupper, ylower)
    plt.sca(axs[n-1])
    # Add a Legend
    plt.legend(fontsize=8)
    # Show the grid lines
    plt.grid(b=True)
    axs[n-1].set_xlabel('r/a')
    axs[n-1].set_ylabel(r'$R_{n,l}(r)$')

# Set the range of x
plt.xlim(xlower,xupper)
plt.ylim(ylower,yupper)



```
