**Exercise**. Draw the first four wave functions and corresponding probability distributions for the particle in a box model. Where is the probability of finding the particle highest and the lowest?


```python
L=1
n_list=np.array([1,2,3,4])
x = np.linspace(0,L, num=200)
```

```python
# Plot the wave functions
fig, axs = plt.subplots(len(n_list),sharex=True, sharey=True,figsize=(4,4))
# Add a title
fig.suptitle('Wavefunction of particle in a 1D box')
for i in range(0,len(n_list)):
    n = n_list[i]
    y = np.sqrt(2/L)*np.sin(n*np.pi*x/L)
    axs[i].plot(x,y, c='b')
    #Plot y=0 as a reference
    axs[i].plot([min(x),max(x)],[0,0],c='k')
    axs[i].text(0.05,0.9,'n='+str(n))
    # Add a grid
    axs[i].grid(alpha=.4,linestyle='--')

# Set the range of x
plt.xlim(0,L)

# Add X and y Label
fig.text(0.5, 0.04, 'x', ha='center')
fig.text(0.01, 0.5, r'$\Psi_n(x)$', va='center', rotation='vertical')


# Add a Legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Plot the probability functions
fig, axs = plt.subplots(len(n_list),sharex=True, sharey=True,figsize=(4,4))
# Add a title
fig.suptitle('Probability density of particle in a 1D box')
for i in range(0,len(n_list)):
    n = n_list[i]
    y = 2/L*np.sin(n*np.pi*x/L)**2
    axs[i].plot(x,y,c='b')
    #Plot y=0 as a reference
    axs[i].plot([min(x),max(x)],[0,0],c='k')
    axs[i].text(0.05,0.9,'n='+str(n))   
    # Add a grid
    axs[i].grid(alpha=.4,linestyle='--')

# Set the range of x
plt.xlim(0,L)
plt.ylim(0,2.5)

# Add X and y Label
fig.text(0.5, 0.04, 'x', ha='center')
fig.text(0.01, 0.5, r'$|\Psi_n(x)|^2$', va='center', rotation='vertical')

# Add a Legend
plt.legend()

# Show the plot
plt.show()
```

```python
epsilon = np.linspace(0,1, num=200)
b=2*np.sqrt(2)*np.pi
```

```python
# size of the figure in inches
plt.figure(figsize=(8,4))

# Plot the rho functions
f = (2*epsilon-1)*np.sin(b*np.sqrt(epsilon)) - 2*np.sqrt(epsilon-epsilon**2)*np.cos(b*np.sqrt(epsilon))
plt.plot(epsilon,f,c='b',label=r'$f(\varepsilon)$')
# Plot the horizontal line as a reference
plt.plot([0,1],[0,0],c='k')
plt.xlim(0,1)

# Add a title
plt.title(r'$f(\varepsilon)$')

# Add X and y Label
plt.ylabel(r'$f(\varepsilon)$')

# Add a grid
plt.grid(alpha=.4,linestyle='--')

# Add a Legend
plt.legend()

# Show the plot
plt.show()
```
